```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el CSV
df = pd.read_csv('drones_dataset.csv')  # Cambia 'drones.csv' por el nombre real de tu archivo

# Mostrar primeras filas del DataFrame
print("Primeras filas del archivo CSV:")
print(df.head())

# Mostrar resumen estad√≠stico
print("\nResumen estad√≠stico:")
print(df.describe(include='all'))

# Visualizar la Distribution de cada columna
for column in df.columns:
    plt.figure(figsize=(8, 4))
    if df[column].dtype in ['int64', 'float64']:
        sns.histplot(df[column], kde=True, bins=20)
        plt.title(f'Distribution of {column}')
    else:
        sns.countplot(data=df, x=column)
        plt.title(f'Frecuencia of values in {column}')
        plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
