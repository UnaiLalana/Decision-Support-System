```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Load the dataset
df = pd.read_csv('drones_dataset.csv')  

print("First rows from the CSV file:")
print(df.head())

# Show the statistical summary
print("\nStatistical summary:")
print(df.describe(include='all'))

# Visualize each column's distribution
for column in df.columns:
    if column != "Drone ID":
        plt.figure(figsize=(8, 4))
        if df[column].dtype in ['int64', 'float64']:
            sns.histplot(df[column], kde=True, bins=20)
            plt.title(f'Distribution of {column}')
        else:
            sns.countplot(data=df, x=column)
            plt.title(f'Value frequency in {column}')
            plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
ids = np.array(df["Drone ID"])

errors = 0

for id in ids:
    sum = np.sum(ids == id)
    if sum != 1:
        print(id + " ID appears more than once: " + str(sum))
        errors += 1

if errors == 0:
    print("Each ID appears only once, therefore each ID is UNIQUE.")

```
